```python
import numpy as np
from scipy.sparse.linalg import minres

np.random.seed(100)
A = np.random.randint(1, 10, (100,100))
A = A + A.T
print(A)
b = np.random.randint(1, 10, 100)
x, exitCode = minres(A, b, maxiter=1000, show=True)
print(exitCode)
np.allclose(A.dot(x), b)
```

```python
import numpy as np
from scipy.sparse.linalg import gmres

np.random.seed(100)
A = np.random.randint(1, 10, (100,100))
A = A + A.T
print(A)
b = np.random.randint(1, 10, 100)

x, exitCode = gmres(A, b, maxiter=1000)
print(exitCode)
np.allclose(A.dot(x), b)
```
